```python
import cv2

video_path = "Videos/Alligator.mp4"
output_folder = "Frames"

cap = cv2.VideoCapture(video_path,0)
count = 0
frame_id = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break
        
    if count % 30 == 0:
        Output_path = f"./frames/frame_{count}.jpg"
        cv2.imwrite(Output_path,frame)
        frame_id += 1

    count += 1

cap.release()
print("Frame extraction completed")
```
