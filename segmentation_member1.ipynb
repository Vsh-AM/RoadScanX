```python
import cv2
import numpy as np
import os

input_folder = r"C:\Users\HM\Desktop\Road ScanX\Frames"
output_folder = r"C:\Users\HM\Desktop\Road ScanX\output"

os.makedirs(output_folder, exist_ok=True)

print("Starting Crack Detection...")

for filename in os.listdir(input_folder):

    if filename.endswith(".jpg") or filename.endswith(".png"):

        path = os.path.join(input_folder, filename)
        image = cv2.imread(path)

        if image is None:
            continue

        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        blur = cv2.GaussianBlur(gray, (9, 9), 0)

        kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (25, 25))
        blackhat = cv2.morphologyEx(blur, cv2.MORPH_BLACKHAT, kernel)

       
        thresh = cv2.adaptiveThreshold(
            blackhat, 255,
            cv2.ADAPTIVE_THRESH_MEAN_C,
            cv2.THRESH_BINARY,
            15, -2
        )

        
        kernel_small = np.ones((3, 3), np.uint8)
        clean = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel_small)

        contours, _ = cv2.findContours(clean, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        output = image.copy()

        for cnt in contours:
            area = cv2.contourArea(cnt)

            if area > 100:   
                cv2.drawContours(output, [cnt], -1, (0, 0, 255), 1)

        
        cv2.imwrite(os.path.join(output_folder, "final_" + filename), output)

        print("Processed:", filename)

print("Detection completed!")


images = [img for img in os.listdir(output_folder) if img.startswith("final_")]
images.sort()

for img_name in images:

    frame = cv2.imread(os.path.join(output_folder, img_name))

    if frame is None:
        continue

    cv2.imshow("Result", frame)

    key = cv2.waitKey(2000)

    if key == ord('q'):
        break

cv2.destroyAllWindows()
```
