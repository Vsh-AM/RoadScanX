```python
import cv2
import numpy as np
from cv2 import cvtColor, COLOR_BGR2RGB, COLOR_RGB2GRAY


video_file_path = 'Transvers.mp4' 

cap = cv2.VideoCapture(video_file_path)
count = 0
f_rate = 30

while cap.isOpened():
    ret, frame = cap.read()
    
    if not ret:
        print("Video Ended")
        break
        
    if count % f_rate == 0:
        
        gray_img = cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
        
        equ_img = cv2.equalizeHist(gray_img)
        
        
        xp = [0, 64, 128, 192, 255]
        fp = [0, 16, 128, 240, 255]
        x = np.arange(256)
        table = np.interp(x, xp, fp).astype('uint8')
        contrast_stretched = cv2.LUT(gray_img, table)

        
        cv2.imwrite(f'Original_{count}.jpg', gray_img)
        cv2.imwrite(f'Equalized_{count}.jpg', equ_img)
        cv2.imwrite(f'Stretched_{count}.jpg', contrast_stretched)
        
        
        combined = np.hstack((gray_img, equ_img, contrast_stretched))
        cv2.imshow('Original | Equalized | Stretched', combined)

    count += 1
    
    if cv2.waitKey(40) == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```
